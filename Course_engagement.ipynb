```python
##In this script we analyse the available data on students and courses 
## To find the most watched courses and
## To find the best reviwed courses
```

```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
student_learning = pd.read_csv("D:/DataScience/365_Learning_Challenge/365_database/365_student_learning.csv")
student_learning.head()
```

```python
course_ratings = pd.read_csv("D:/DataScience/365_Learning_Challenge/365_database/365_course_ratings.csv")
course_ratings.head()
```

```python
## extraction of course_id and minutes watched columns
course_minutes =student_learning[['course_id','minutes_watched']].groupby('course_id', as_index=False)['minutes_watched'].sum() # Total minutes
course_average_minutes =student_learning[['course_id','minutes_watched']].groupby('course_id',as_index=False)['minutes_watched'].mean()
```

```python
df_minutes = pd.DataFrame(course_minutes)
df_avg_minutes = pd.DataFrame(course_average_minutes)
```

```python
df_minutes['average_minutes'] = df_avg_minutes['minutes_watched']
df_minutes
```

```python
top_courses = df_minutes.sort_values(by=['average_minutes'], ascending=False).head()
top_courses #top courses by average minutes watched
```

```python
## Creation of a dictionary of functions
funcs = {'average_course_rating': 'mean', 'no_of_ratings': 'count'}
```

```python
## Rating analysis on the different courses
course_ratings_grouped =course_ratings[['course_id','course_rating']].groupby('course_id', as_index=False)['course_rating'].agg(funcs)
course_ratings_grouped =pd.DataFrame(course_ratings_grouped) ## converting the groupby object to a dataframe
course_ratings_grouped.head()
```

```python
## concatenating the minutes and ratings database
##df_minutes, course_ratings_grouped
df_final = pd.merge(left=df_minutes, right= course_ratings_grouped, on= 'course_id',
                   left_index=False, right_index=False)
df_final.head(10)
```

```python
## sorting the db by minutes watched
df_sorted = df_final.sort_values(by='minutes_watched', axis=0, ascending = False, ignore_index= True)
df_sorted
```

```python
## Writing the dataset to a CSV File
df_final.to_csv('course_stats.csv', encoding='utf-8', index=False)
```
